```python
def top_counts(df, column, n=10):
    return df[column].value_counts().head(n)

top_counts(spotify, "Artist")
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_top_albums_by_likes(df, n=10):
    top_albums = df.groupby("Album")["Likes"].sum().sort_values(ascending=False).head(n)
    top_albums.plot(kind="bar")
    plt.xlabel("Album")
    plt.ylabel("Total Likes")
    plt.title("Top Albums by Likes")
    plt.show()

plot_top_albums_by_likes(spotify)
```

Which albums audience-ku romba pidichirukku nu theriyum.

```python
def plot_scatter(df, x, y):
    plt.scatter(df[x], df[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"{x} vs {y}")
    plt.show()

plot_scatter(spotify, "Stream", "Likes")
```


Streams increase aagum pothu Likes-um increase aagutha nu check pannrom.
